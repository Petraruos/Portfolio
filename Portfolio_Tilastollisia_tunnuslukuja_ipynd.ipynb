```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('seaborn-dark')
```

```python
df = pd.read_excel('http://taanila.fi/kunnat.xlsx')
df.head()
```

```python
df.describe().style.format('{:.2f}')
```

```python
df[['Väkiluku, 2018', 'Alle 15-vuotiaiden osuus väestöstä, %, 2018', '15-64 -vuotiaiden osuus väestöstä, %, 2018', 'Yli 64-vuotiaiden osuus väestöstä, %, 2018']].describe().style.format('{:.2f}')
```

```python
df1 = df[['Väkiluku, 2018', 'Alle 15-vuotiaiden osuus väestöstä, %, 2018', '15-64 -vuotiaiden osuus väestöstä, %, 2018', 'Yli 64-vuotiaiden osuus väestöstä, %, 2018']].describe()
df1.index = ['lukumäärä', 'keskiarvo', 'keskihajonta', 'pienin', 'alaneljännes', 'mediaani', 'yläneljännes', 'suurin']
df1.style.format('{:.2f}')
```

```python
df2 = df[['Vuosikate, euroa/asukas, 2017', 'Lainakanta, euroa/asukas, 2017', 'Konsernin lainakanta, euroa/asukas, 2017']].max().to_frame('suurin')
df2
```

```python
ax1 = df2.sort_values('suurin').plot.barh(legend=False)
ax1.set_xlim(1000,13000)
ax1.set_xlabel('Suurimmat vuosikatteet ja lainat asukasta kohden')
```
