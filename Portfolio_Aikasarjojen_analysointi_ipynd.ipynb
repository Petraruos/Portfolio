```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('seaborn-pastel')

from matplotlib.ticker import MultipleLocator
from matplotlib.dates import DateFormatter

myLoc = MultipleLocator(14)

myFmt = DateFormatter('%d/%m')
```

```python
df=pd.read_excel('http://taanila.fi/CO2.xlsx')
df.index = pd.to_datetime(df['Kuukausi'],format="%Y-%m")
df=df.drop('Kuukausi',axis=1)
df.head()
```

```python
df.plot()
```

```python
df.resample('Q').mean().plot()
```

```python
df.resample('Y').mean().plot()
```

```python
df.resample('Q').sum().plot()
```

```python
df['CO2'].plot()

df['CO2'].rolling(50).mean().plot()
df['CO2'].rolling(200).mean().plot()
```

```python
df['CO2%'] = df['CO2'].pct_change()
```

```python
muutokset = pd.concat([df['CO2%']], axis=1)
muutokset.head()
```

```python
ax2 = muutokset['2019-02-01':].plot()

ax2.set_ylabel('Muutos')
```
