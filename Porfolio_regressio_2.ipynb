```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_excel('http://taanila.fi/Concrete_Data.xlsx')


df.head()
```

```python
df['age2'] = np.log(df['Age (day)'])
df.head()
```

```python
df.isnull().sum()
```

```python
sns.distplot(df['Cement (component 1)(kg in a m^3 mixture)'], bins=30)
```

```python
correlation_matrix = df.corr().round(2)

plt.figure(figsize=(12,9))

sns.heatmap(data=correlation_matrix, annot=True)
```

```python
features = ['Age (day)', 'Superplasticizer (component 5)(kg in a m^3 mixture)']

target = df['Cement (component 1)(kg in a m^3 mixture)']

plt.figure(figsize=(15, 5))

for i, feature in enumerate(features):
    plt.subplot(1, len(features), i+1)
    plt.scatter(df[feature], target)
    plt.xlabel(feature)
    plt.ylabel('Cement (component 1)(kg in a m^3 mixture)')
```
